```python
import gensim.downloader as api

word_vectors = api.load("glove-wiki-gigaword-100")
words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']
word_embeddings = {word: word_vectors[word] for word in words}
for word, vector in word_embeddings.items():
    print(f"Word: {word}, Embedding: {vector[:10]}...")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
word_vectors_pca = pca.fit_transform(list(word_embeddings.values()))
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    plt.scatter(word_vectors_pca[i, 0], word_vectors_pca[i, 1])
    plt.text(word_vectors_pca[i, 0] + 0.01, word_vectors_pca[i, 1] + 0.01, word, fontsize=9)
plt.show()
```
